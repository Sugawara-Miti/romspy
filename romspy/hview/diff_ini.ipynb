```python
# coding: utf-8
# (c) 2016-02-10 Teruhisa Okada

%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np
import romspy
import hview

inifile = '/home/okada/ism-i/apps/OB500P/testDA/param5-001/output/ob500_ini_24.nc'
grdfile = '/home/okada/Data/ob500_grd-11_3.nc'
mapfile = '/home/okada/romspy/romspy/deg_OsakaBayMap_okada.bln'

nc = hview.Dataset(inifile, grdfile, mapfile)
nc.check_time()
```

```python
romspy.cmap('OrRd')
#romspy.cmap('jet')

kw = {}
kw['levels'] = np.arange(10,32,2)
kw['label'] = r'Temperature ($\degree$C)'

def diff_ini(nc, vname, k, **kw):
    fig, ax = plt.subplots(1,2,figsize=(13,5))
    plt.subplot(1,2,1)
    nc.sview(vname, t=0, k=k, **kw)
    plt.subplot(1,2,2)
    nc.sview(vname, t=1, k=k, **kw)

diff_ini(nc, 'temp', 19, **kw)
```

```python
romspy.cmap('Blues')

kw = {}
kw['levels'] = np.arange(24,33.1,1)
kw['label'] = 'Salinity'

diff_ini(nc, 'salt', 19, **kw)
```

```python
romspy.cmap('YlGn')

kw = {}
kw['levels'] = np.arange(1,10.5,1)
kw['label'] = 'Chlorophyll ($\mu$g/l)'

diff_ini(nc, 'chlorophyll', 19, **kw)
```

```python
romspy.cmap('BuPu_r')

kw = {}
kw['cff'] = romspy.mol2g_O2
kw['levels'] = np.arange(1,11,1)
kw['label'] = 'DO (mg/l)'

diff_ini(nc, 'oxygen', 19, **kw)
```
