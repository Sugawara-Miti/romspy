```python
# coding: utf-8
# (c) 2016-02-10 Teruhisa Okada

%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np
import romspy
import hview

inifile = '/home/okada/ism-i/apps/OB500P/testDA/param5-001/output/ob500_ini_24.nc'
grdfile = '/home/okada/Data/ob500_grd-11_3.nc'
mapfile = '/home/okada/romspy/romspy/deg_OsakaBayMap_okada.bln'

nc = hview.Dataset(inifile, grdfile, mapfile)
nc.check_time()
```

```python
t = 0

romspy.cmap('OrRd')
#romspy.cmap('jet')

kw = {}
kw['levels'] = np.arange(10,32,1)
kw['label'] = r'Temperature ($\degree$C)'

def main(nc, vname, t, **kw):
    fig, ax = plt.subplots(1,2,figsize=(13,5))
    plt.subplot(1,2,1)
    nc.sview(vname, t=t, k=19, **kw)
    plt.subplot(1,2,2)
    nc.sview(vname, t=t, k=0, **kw)

main(nc, 'temp', t, **kw)
```

```python
romspy.cmap('Blues')

kw = {}
kw['levels'] = np.arange(24,33.1,1)
kw['label'] = 'Salinity'

main(nc, 'salt', t, **kw)
```

```python
romspy.cmap('YlGn')

kw = {}
kw['levels'] = np.arange(1,10.5,1)
kw['label'] = 'Chlorophyll ($\mu$g/l)'

main(nc, 'chlorophyll', t, **kw)
```

```python
romspy.cmap('BuPu_r')

kw = {}
kw['cff'] = romspy.mol2g_O2
kw['levels'] = np.arange(1,11,1)
kw['label'] = 'DO (mg/l)'

main(nc, 'oxygen', t, **kw)
```
